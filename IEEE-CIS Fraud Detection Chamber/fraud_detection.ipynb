##Reduce Memory Usage

```python
def reduce_mem_usage(df, verbose=True):
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    start_mem = df.memory_usage().sum() / 1024**2
    for col in df.columns:
        col_type = df[col].dtypes
        if col_type in numerics:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
    end_mem = df.memory_usage().sum() / 1024**2
    if verbose: print('Mem. usage decreased to {:5.2f} Mb ({:.1f}% reduction)'.format(end_mem, 100 * (start_mem - end_mem) / start_mem))
    return df
```

```python
identity_df= reduce_mem_usage(identity_df)
transaction_df= reduce_mem_usage(transaction_df)
```

##Data Cleaning
###Find and remove columns that do not have corresponing transaction ID info

```python
# Merge transaction_df and identity_df on transactionID (inner join)
merged_df = transaction_df.merge(identity_df, how='inner', on='TransactionID')

# Now merged_df contains entries with corresponding IDs in both DataFrames

# Select desired columns from merged_df for each DataFrame
clean_transaction_df = merged_df[transaction_df.columns]
clean_identity_df = merged_df[identity_df.columns]
print(clean_identity_df.shape)
print(clean_transaction_df.shape)
```

###Find and Drop columns with more that 70% Null Values

```python
#Define function that will be used to calculate column null percentage
def get_null_percentage(df, threshold=70):
    null_counts = df.isnull().sum()
    null_percentages = (null_counts / len(df)) * 100
    return null_percentages[null_percentages > threshold].index.tolist()
```

```python
#Calculate for both dataframes
#Calling the list df_sus_cols - because  they are suspicious; why do they have so many NaN
identity_sus_cols = get_null_percentage(clean_identity_df)
transaction_sus_cols = get_null_percentage(clean_transaction_df)
print(identity_sus_cols)
print(transaction_sus_cols)
```

```python
#Drop the columns with >70% null values
clean_identity_df.drop(columns = identity_sus_cols, inplace=True)
clean_transaction_df.drop(columns = transaction_sus_cols, inplace=True)
```

###Impute missing int values

```python
num_id_cols = clean_identity_df.select_dtypes(include=['int']).columns
num_trans_cols = clean_transaction_df.select_dtypes(include=['int']).columns


for col in num_id_cols:
  clean_identity_df[col].fillna(clean_identity_df[col].mean(), inplace=True)

for col in num_trans_cols:
  clean_transaction_df[col].fillna(clean_transaction_df[col].mean(), inplace=True)
```

```python
df_train = clean_transaction_df.merge(clean_identity_df, how='inner', on='TransactionID')
df_train.head()
```

###Impute Categorical Data

```python
def get_unique_counts_for_categorical_data(df):
  """
  This function analyzes categorical columns in a DataFrame and returns a dictionary
  containing the number of unique values for each column.

  Args:
      df (pandas.DataFrame): The DataFrame containing the data.

  Returns:
      dict: A dictionary where keys are column names and values are the number of unique items.
  """
  categorical_cols = df.select_dtypes(include=['object']).columns
  unique_counts = {}
  for col in categorical_cols:
    unique_counts[col] = df[col].nunique()  # Count unique values using nunique()
  return unique_counts

unique_counts = get_unique_counts_for_categorical_data(df_train)

for col, count in unique_counts.items():
  print(f"Column '{col}': {count} unique items")
```
